```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

### Upload Data with Requests and BeautifulSoup

```python
page= requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
page
```

The BeautifulSoup Package is used to read the table from the link above

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
table = soup.find('table')
table
```

```python
column_names = ['Postalcode','Borough','Neighborhood']
df = pd.DataFrame(columns = column_names)
```

```python
for tr_cell in table.find_all('tr'):
    row_data=[]
    for td_cell in tr_cell.find_all('td'):
        row_data.append(td_cell.text.strip())
    if len(row_data)==3:
        df.loc[len(df)] = row_data
```

```python
df.head()
```

Drop Cells That Do Not Have An Assigned Borough

```python
df_filtered=df[df['Borough'] != 'Not assigned']
```

```python
df_filtered
```

```python
df_f2 = df_filtered.reset_index(0)
df_f2
```

```python
df_final = df_f2.drop(['index'], axis =1)
df_final
```

```python
df_final.shape
```

```python
import matplotlib.cm as cm
import matplotlib.colors as colors
from geopy.geocoders import Nominatim

from sklearn.cluster import KMeans
import folium
import numpy as np
```

```python
address = 'Toronto, ON'
geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
```

```python
# create map of New York using latitude and longitude values
map_Toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_zip['Latitude'], df_zip['Longitude'], df_zip['Borough'], df_zip['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto)  
    
map_Toronto
```
